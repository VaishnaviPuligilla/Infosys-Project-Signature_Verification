```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import numpy as np
import glob

# Set parameters
image_size = 256
batch_size = 32
channel = 3

# Define file paths for the dataset
real_sign = (
    glob.glob(r"D:\New_Dataset\Dataset\Real\real\*.*")
)

forg_sign = (
    glob.glob(r"D:\New_Dataset\Dataset\Forge\forge\*.*")
)

# Directory structure for loading images using image_dataset_from_directory
dataset_directory = "D:/New_Dataset/Dataset"

# Load training dataset
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_directory,  # Root directory of dataset
    image_size=(image_size, image_size),
    batch_size=batch_size,
    label_mode='int',  # Use integer labels for classification
    shuffle=True,
    validation_split=0.2,  # Split 80% for training, 20% for validation
    subset='training',  # Use 'training' subset for training data
    seed=123  # Set a seed for reproducibility
)

# Create validation dataset from the same directory
val_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_directory,  # Root directory of dataset
    image_size=(image_size, image_size),
    batch_size=batch_size,
    label_mode='int',
    shuffle=True,
    validation_split=0.2,
    subset='validation',  # Use 'validation' subset for validation data
    seed=123
)

# Get class names
class_names = train_dataset.class_names
print("Class Names:", class_names)

# Display sample images
plt.figure(figsize=(20, 10))
for image_batch, label_batch in train_dataset.take(1):  
    for i in range(12):  
        ax = plt.subplot(3, 4, i + 1)
        plt.imshow(image_batch[i].numpy().astype('uint8'))
        plt.title(class_names[label_batch[i]])
        plt.axis('off')

# Define the model with explicit input_shape in an InputLayer
input_shape = (image_size, image_size, channel)
n_classes = len(class_names)

model = models.Sequential([
    layers.InputLayer(input_shape=input_shape),   # Explicitly add an InputLayer
    layers.Rescaling(1./255),  # Rescale images to [0, 1] range
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(n_classes, activation='softmax')  # Output layer for classification
])

# Compile the model
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)

# Check model summary
model.summary()

# Train the model
epochs = 50
history = model.fit(
    train_dataset,  # Training dataset
    epochs=epochs,
    validation_data=val_dataset,  # Validation dataset
)

# Save the trained model
model.save('signature_verification_model.keras')

# Plot training history
acc_his = history.history['accuracy']
val_acc_his = history.history['val_accuracy']
loss_his = history.history['loss']
val_loss_his = history.history['val_loss']

plt.figure(figsize=(20, 8))
ax = plt.subplot(1, 2, 1)
plt.plot(range(epochs), acc_his, label='Training Accuracy')
plt.plot(range(epochs), val_acc_his, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

ax = plt.subplot(1, 2, 2)
plt.plot(range(epochs), loss_his, label='Training Loss')
plt.plot(range(epochs), val_loss_his, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()

# Make predictions on a test batch
for image_batch, label_batch in val_dataset.take(1):
    first_image = image_batch[0].numpy().astype('uint8')
    first_label = label_batch[0].numpy()
    print('First image to predict')
    plt.imshow(first_image)
    print('Actual label:', class_names[first_label])
    batch_predictions = model.predict(image_batch)
    print('Predicted label:', class_names[np.argmax(batch_predictions[0])])

# Prediction function
def predict(model, img):
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)
    predictions = model.predict(img_array)
    predicted_class = class_names[np.argmax(predictions[0])]
    confidence = round(100 * (np.max(predictions[0])), 2)
    return predicted_class, confidence

# Plot predictions
plt.figure(figsize=(20, 8))
for images, labels in val_dataset.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype('uint8'))
        predicted_class, confidence = predict(model, images[i].numpy())
        actual_class = class_names[labels[i]]
        plt.title(f'Actual: {actual_class},\nPredicted: {predicted_class},\nConfidence: {confidence}%')
        plt.axis('off')
plt.show()
```

```python
# Assuming val_dataset is already defined in the previous steps

# Evaluate the model on the validation dataset (or test dataset)
final_loss, final_accuracy = model.evaluate(val_dataset)
print('Final loss: {:.2f}%'.format(final_loss * 100))
print('Final accuracy: {:.2f}%'.format(final_accuracy * 100))
```
